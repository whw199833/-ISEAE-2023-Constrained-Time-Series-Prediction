```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
td = pd.read_csv("data/test_data.csv")
td
```

```python
data=td.groupby(['vehicle_id','construction_end_time'])['repair_item_code'].apply(lambda x:x.astype(str).str.cat(sep=',')).reset_index()
data['construction_end_time'] = pd.to_datetime(data['construction_end_time'])
data_asc=data.sort_values(['vehicle_id', 'construction_end_time'],ascending=[True, True])
data_asc
```

```python
data2=data_asc.groupby('vehicle_id')['repair_item_code'].apply(lambda x:x.str.cat(sep=' ')).reset_index()
data2
```

```python
# 生成车辆ID和车型的字典
id2model_map=td.set_index('vehicle_id')['vehicle_model_code'].to_dict()
id2model_map
```

```python
data2['vehicle_id']=data2['vehicle_id'].map(id2model_map)
data2
```

```python
a = data2['vehicle_id'].astype(str)
b = data2['repair_item_code'].str.rsplit(' ').str[:-1].str.join(',')
c = data2['repair_item_code'].str.rsplit(' ').str[-1]
d = a.str.cat(b,sep='\t').str.cat(c,sep='\t')# 用制表符 \t 将最后一次维修记录跟之前的维修记录分隔开
# c = c[~c.str.startswith('\t')]# 删除只有一次维修记录的数据
d
```

```python
# c.to_csv('data/train_data.txt',sep=' ',index=0,header=0)
d.to_csv('data/test_data_addfeature.txt',sep=' ',index=0,header=0)
```
