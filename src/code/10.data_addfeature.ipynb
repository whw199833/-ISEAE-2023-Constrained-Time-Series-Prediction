```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
td = pd.read_csv("data/all_data.csv")
td
```

```python
data = td[td['repair_item_code'] != 100197]
td = data[data['repair_item_code'] != 100009]
td
```

```python
data = td.groupby(['repair_item_name'])['repair_item_name'].agg('count').to_frame('d').reset_index()
data
dd = dict(zip(data.repair_item_name, data.d))
cc = 0
for i in dd:
    if dd[i] < 10000 and dd[i] > 1000:
        print(dd[i])
        print(i)
        print("===")
        cc += 1
#print(cc)
dd
```

```python
vehicle_model_code_vehicle_id_dict = dict(zip(td.vehicle_id, td.vehicle_model_code))
vehicle_model_code_vehicle_id_dict
```

```python
data=td.groupby(['vehicle_model_code']).count().reset_index()
data = data.vehicle_model_code
data.index = data.index+1# 索引从1开始 
# 把series格式转为dataframe格式
data = pd.DataFrame(data)
# 生成字典vehicle_model_code字典
vehicle_model_dic=data.set_index(data.index)['vehicle_model_code'].to_dict()
vehicle_model_dic
area_dict = dict(zip(data.index, data.vehicle_model_code))
area_dict_re = dict(zip(data.vehicle_model_code,data.index))
area_dict_re
```

```python
data = td.groupby(['vehicle_id','vehicle_model_code']).agg('count').reset_index()
data = data.groupby(['vehicle_model_code'])['vehicle_model_code'].agg('count').to_frame('d').reset_index()
max_value = data['d'].max()
data['d']=data['d']/max_value
vehicle_model_code_num_dict = dict(zip(data.vehicle_model_code, data.d))
vehicle_model_code_num_dict
```

```python
data = td.groupby(['vehicle_id','construction_id'])['vehicle_id'].agg('count').to_frame('c').reset_index()
data = data.groupby(['vehicle_id'])['vehicle_id'].agg('count').to_frame('c').reset_index()
vehicle_construction_num_dict = dict(zip(data.vehicle_id, data.c-1))# 总次数-1次
vehicle_construction_num_dict
```

```python
data = td.groupby(['vehicle_id','construction_id'])['mileage'].agg('max').to_frame('c').reset_index()
data = data.groupby(['vehicle_id'])['c'].agg('max').to_frame('d').reset_index()
data
max_value = data['d'].max()
data['d']=data['d']/max_value
vehicle_mileage_dict = dict(zip(data.vehicle_id, data.d))
vehicle_mileage_dict
```

```python
data = td.groupby(['vehicle_id','repair_item_code'])['vehicle_id'].agg('count').to_frame('c').reset_index()
data = data.groupby(['vehicle_id'])['c'].agg('sum').to_frame('c').reset_index()
vehicle_repair_item_dict = dict(zip(data.vehicle_id, data.c))
vehicle_repair_item_dict
```

```python
data=td.groupby(['vehicle_id','construction_end_time'])['repair_item_code'].apply(lambda x:x.astype(str).str.cat(sep=',')).reset_index()
data['construction_end_time'] = pd.to_datetime(data['construction_end_time'])
data_asc=data.sort_values(['vehicle_id', 'construction_end_time'],ascending=[True, True])
data_asc
```

```python
data2=data_asc.groupby('vehicle_id')['repair_item_code'].apply(lambda x:x.str.cat(sep=' ')).reset_index()
data2
data2['vehicle_repair_item'] = data2.apply(lambda x: vehicle_repair_item_dict[x['vehicle_id']], axis=1)
data2['vehicle_mileage'] = data2.apply(lambda x: vehicle_mileage_dict[x['vehicle_id']], axis=1)
data2['vehicle_construction_num'] = data2.apply(lambda x: vehicle_construction_num_dict[x['vehicle_id']], axis=1)
data2['vehicle_model_code_num'] = data2.apply(lambda x: vehicle_model_code_num_dict[vehicle_model_code_vehicle_id_dict[x['vehicle_id']]], axis=1)
data2['vehicle_model_code_id'] = data2.apply(lambda x: area_dict_re[vehicle_model_code_vehicle_id_dict[x['vehicle_id']]], axis=1)
data2
```

```python
a = data2['vehicle_model_code_id'].astype(str)
b = data2['vehicle_model_code_num'].astype(str)
c = data2['vehicle_mileage'].astype(str)
d = data2['vehicle_construction_num'].astype(str)
e = data2['vehicle_repair_item'].astype(str)
f = data2['repair_item_code'].str.rsplit(' ').str[:-1].str.join(',')
g = data2['repair_item_code'].str.rsplit(' ').str[-1]
h = a.str.cat(b,sep=',').str.cat(c,sep=',').str.cat(d,sep=',').str.cat(e,sep=',').str.cat(f,sep='\t').str.cat(g,sep='\t')# 用制表符 \t 将最后一次维修记录跟之前的维修记录分隔开

# f = data2['repair_item_code']
# h = a.str.cat(b,sep=',').str.cat(c,sep=',').str.cat(d,sep=',').str.cat(e,sep=',').str.cat(f,sep='\t')# 用制表符 \t 将最后一次维修记录跟之前的维修记录分隔开
h
```
