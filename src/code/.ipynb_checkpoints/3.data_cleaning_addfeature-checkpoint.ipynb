```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
md = pd.read_csv("data/train_data.csv")
md
```

```python
md.shape
```

```python
## 数据缺失值可视化
import missingno as msno
msno.bar(md)
```

```python
#删除无用的列
md = md.drop(['creater_id','create_time_x','updater_id','update_time_x','construction_no_x','construction_supplier_type','construction_garage_no','has_change_repair_items','report_description','approval_id','approval_status','approval_had_reject','reject_number','vehicle_category_code','vehicle_category_name','appraise_submited','appraise_submit_time','is_offsite','is_cooperation','id_y','create_time_y','update_time_y','construction_no_y','parts_type','construction_city_code','construction_city_name','contruction_total_fee_amount','maintenance_last_time','maintenance_last_mileage'],axis=1)
md
```

```python
# 将车辆只有一次维修记录的数据删除
grouped=md.groupby(['vehicle_id','construction_end_time'])['repair_item_code'].apply(lambda x:x.astype(str).str.cat(sep=',')).reset_index()
grouped = grouped.groupby('vehicle_id').filter(lambda x: len(x) > 1)
vehicle_ids = grouped.get('vehicle_id')
vehicle_ids
md = md[md['vehicle_id'].isin(vehicle_ids)]
md.to_csv("data/all_data.csv")
md
```

```python
# 将数据拆分成训练集和测试集，占比为：8:2

# 对 column_name 列进行拆分，将 80% 的数据作为训练集，20% 的数据作为测试集
vehicles = md.groupby('vehicle_id')['vehicle_id'].first()
train_vehicles = vehicles.sample(frac=0.8, random_state=0)
test_vehicles = vehicles.drop(train_vehicles.index)
# 输出训练集和测试集的大小
print(f'Training data size: {len(train_vehicles)}')
print(f'Testing data size: {len(test_vehicles)}')
train_data = md[md['vehicle_id'].isin(train_vehicles)]
test_data = md[md['vehicle_id'].isin(test_vehicles)]
# 保存数据
train_data.to_csv('data/train_data.csv')
test_data.to_csv('data/test_data.csv')
```

```python
# 以下运行两次，得到训练集和测试集
# data_type = train_data
data_type = test_data
data=data_type.groupby(['vehicle_id','construction_end_time'])['repair_item_code'].apply(lambda x:x.astype(str).str.cat(sep=',')).reset_index()
data['construction_end_time'] = pd.to_datetime(data['construction_end_time'])
data_asc=data.sort_values(['vehicle_id', 'construction_end_time'],ascending=[True, True])
data_asc
```

```python
data2=data_asc.groupby('vehicle_id')['repair_item_code'].apply(lambda x:x.str.cat(sep=' ')).reset_index()
data2
```

```python
#删除车辆ID
data_ = data2.drop('vehicle_id', axis=1)
data_
```

```python
a = data_['repair_item_code'].str.rsplit(' ').str[:-1].str.join(',')
b = data_['repair_item_code'].str.rsplit(' ').str[-1]
c = a.str.cat(b,sep='\t')# 用制表符 \t 将最后一次维修记录跟之前的维修记录分隔开
# c = c[~c.str.startswith('\t')]# 删除只有一次维修记录的数据
c
```

```python
# c.to_csv('data/train_data.txt',sep=' ',index=0,header=0)
c.to_csv('data/test_data.txt',sep=' ',index=0,header=0)
```
